```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_l = 32
y_l = x_l

x_positions = np.arange(x_l) * 2*np.pi / x_l
y_positions = x_positions

x_modes = 32
y_modes = x_modes
```

```python
V_x = np.zeros([x_modes *2, x_l], dtype=np.cfloat)
for row in range(x_modes):
     for col in range(x_l):
        V_x[row, col] = np.exp(-1j * row *  x_positions[col]) 
        V_x[-(row+1), col] = np.exp(-1j * (x_l - row) *  x_positions[col]) 
V_x = np.divide(V_x, np.sqrt(x_l))


V_y = np.zeros([y_modes *2, y_l], dtype=np.cfloat)
for row in range(y_modes):
     for col in range(y_l):
        V_y[row, col] = np.exp(-1j * row *  y_positions[col]) 
        V_y[-(row+1), col] = np.exp(-1j * (y_l - row) *  y_positions[col]) 
V_y = np.divide(V_y, np.sqrt(y_l))
```

```python
# x,y = np.meshgrid(x_positions, y_positions)
x,y = np.mgrid[0:64,0:64]
out = np.matmul(V_x, np.conj(np.transpose(V_x)))

plt.contourf(x,y,out)
```
